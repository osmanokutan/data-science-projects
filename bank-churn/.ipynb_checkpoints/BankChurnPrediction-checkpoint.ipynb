```python
#to include Gender feature into correlation analysis, we encode it by by Male -> 0, Female->1
gender_dic = {'Male':0, 'Female':1}
data['Gender']=data['Gender'].apply(lambda s: gender_dic.get(s,s))

plt.figure(figsize=(9,6))
corr_matrix=data.corr()
mask = np.zeros_like(corr_matrix, dtype=np.bool_)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(
    corr_matrix,
    center=0,
    annot=True,
    fmt=".2f",
    linewidth=1,
    mask=mask
);
```

The input features are mostly uncorrelated, and some of them have high correlations with the target feature, especially Age, NumOfProducts and IsActiveMember.

# Encoding

```python
#one hot encode the Geography feature
data=pd.get_dummies(data)
```

```python
data.info()
```

# Classification

We will train logistic regression, decision tree, random forest, and xgboost classifiers. We hypertune them using cross validation. Our main performence metric will be F1-score to address class imbalance. We will also keep track of accuracy, precision, recall, and roc-auc scores.

```python
X=data.drop('Exited',axis=1)
y=data['Exited']
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,shuffle=True, random_state=13)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
metrics=['f1','recall','precision','accuracy','roc_auc']
score_features=['mean_test_'+metric for metric in metrics]
def plot_cv(results, metrics, params, name, emph, ticks=False):
    features=['mean_test_'+metric for metric in metrics]
    for metric, feature in zip(metrics, features):
        if metric==emph:
            plt.plot(params,results[feature],label=metric,linewidth=3)
        else:
            plt.plot(params,results[feature],label=metric)
        plt.xlabel(name)
        plt.ylabel('Mean CV score')
        if ticks:
            plt.xticks(params)
    best_index=results['mean_test_'+emph].argmax()
    best_param=params[best_index]
    best_score=results['mean_test_'+emph].max()
    xs=[best_param]*10
    ys=np.linspace(0,best_score,10)
    plt.plot(xs,ys,linestyle='dashed')
    plt.annotate("%0.4f" % best_score, (best_param, best_score + 0.005))
    plt.legend()
    return results[features].iloc[best_index]
```

## Logistic Regression

```python
logistic_params = {'class_weight':[{0:p,1:1-p} for p in np.linspace(0,1,21)]}
best_lr = GridSearchCV(
    estimator=LogisticRegression(random_state=13), 
    param_grid=logistic_params, 
    cv=5,
    scoring=metrics,
    refit='f1',
    verbose=1
)

best_lr.fit(X_train_scaled, y_train);
```

```python
print('Best CV parameters for logistic regression:',best_lr.best_params_)
print('Best CV f1-score for logistic regression:',f'{best_lr.best_score_:.4f}')
```

```python
results=pd.DataFrame(best_lr.cv_results_)
plt.figure(figsize=(8,8))
ps=np.linspace(0,1,21)
lr_scores=plot_cv(results,metrics,ps,'Class-0 weight',emph='f1')
```

## Decision Tree

```python
tree_params = {'max_leaf_nodes':range(2,100)}
best_dt = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=13,class_weight={0:0.3,1:0.7}), 
    param_grid=tree_params, 
    cv=5,
    scoring=metrics,
    refit='f1',
    verbose=1
)

best_dt.fit(X_train, y_train);
```

```python
print('Best CV parameters for decision tree:',best_dt.best_params_)
print('Best CV f1-score for decision tree:',f'{best_dt.best_score_:.4f}')
```

```python
plt.figure(figsize=(8,8))
results=pd.DataFrame(best_dt.cv_results_)
params=list(range(2,100))
dt_scores=plot_cv(results,metrics,params,'Max Leaf Nodes',emph='f1')
```

## Random Forest

```python
forest_params = {'max_depth':range(2,13),'max_features':range(4,8)}
best_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=13,class_weight={0:0.3,1:0.7}), 
    param_grid=forest_params, 
    cv=5,
    scoring=metrics,
    refit='f1',
    verbose=1
)

best_rf.fit(X_train, y_train);
```

```python
print('Best CV parameters for random forest:',best_rf.best_params_)
print('Best CV f1-score for random forest:',f'{best_rf.best_score_:.4f}')
```

```python
plt.figure(figsize=(8,8))
results=pd.DataFrame(best_rf.cv_results_)
I=results['param_max_features']==6
params=list(range(2,13))
rf_scores=plot_cv(results[I],metrics,params,'Max Depth',emph='f1',ticks=True)
```

## XGBoost

```python
boost_params = {'n_estimators':range(20,220,20), 'learning_rate':np.linspace(0.01,1,10)}
best_xgb = GridSearchCV(
    estimator=XGBClassifier(random_state=13,class_weight={0:0.3,1:0.7}), 
    param_grid=boost_params, 
    cv=5,
    scoring=metrics,
    refit='f1',
    verbose=1
)

best_xgb.fit(X_train, y_train);
```

```python
print('Best CV parameters for XGBoostClassifier:',best_xgb.best_params_)
print('Best CV f1-score for XGBoostClassifier',f'{best_xgb.best_score_:.4f}')
```

```python
plt.figure(figsize=(8,8))
results=pd.DataFrame(best_xgb.cv_results_)
I=results['param_n_estimators']==140
params=np.linspace(0.01,1,10)
xgb_scores=plot_cv(results[I],metrics,params,'Learning rate',emph='f1',ticks=True)
```

# Model Comparison and Selection

```python
scores = pd.DataFrame(np.array([lr_scores,dt_scores, rf_scores, xgb_scores]),
                      index=['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost'],
                      columns=['F1 CV','Recall CV','Precision CV', 'Accuracy CV', 'ROC-AUC CV']
)
scores.sort_values(by='F1 CV',ascending=False)
```

We choose Random Forest as our final model, since it has the highest cross validation F1 score, and its performance is always in top two in other metrics.

# Test Scores

```python
rf=RandomForestClassifier(max_depth=9, max_features=5, class_weight={0:0.3,1:0.7}, random_state=13)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
y_proba=rf.predict_proba(X_test)[:,1]
```

```python
print('Test F1-score:',f'{f1_score(y_test,y_pred):.4f}')
print('Test accuracy:',f'{accuracy_score(y_test,y_pred):.4f}')
print('Test precision:',f'{precision_score(y_test,y_pred):.4f}')
print('Test recall:',f'{recall_score(y_test,y_pred):.4f}')
print('Test roc-auc-score:',f'{roc_auc_score(y_test,y_proba):.4f}')
```

Let us see the feature importances assigned by the random forest.

```python
importances=pd.DataFrame(
    rf.feature_importances_,
    index=X_train.columns,
    columns=['Feature importance']
)
importances.sort_values(by='Feature importance',ascending=False)
```
